```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36"
options = webdriver.chrome.options.Options()
options.headless = False
options.add_argument(f'user-agent={user_agent}')
options.add_argument("--window-size=1920,1080")
options.add_argument('--ignore-certificate-errors')
options.add_argument('--allow-running-insecure-content')
options.add_argument("--disable-extensions")
options.add_argument("--proxy-server='direct://'")
options.add_argument("--proxy-bypass-list=*")
options.add_argument("--start-maximized")
options.add_argument('--disable-gpu')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument("--window-size=500,1000")
```

```python
def adam_scrapping(no_of_option):
    driver = webdriver.Chrome(options=options)
    driver.get("https://www.cmegroup.com/markets/fx/g10/british-pound.settlements.html")
    
    options_dropdown_click2 = driver.find_element(By.XPATH, option_dropdown_list[no_of_option]).click()
    expiration_dropdown_click2 = WebDriverWait(driver, 10).until(ec.visibility_of_element_located((By.XPATH, expiration_dropdown_list[no_of_option]))).click()
    button_load_all = driver.find_element(By.XPATH,"//button[@rel='noopener noreferrer']").click()
    
    global strike_series
    global put_prior_oi_series
    global put_est_vol_series 
    global put_oi_series 
    global call_prior_oi_series 
    global call_est_vol_series
    global call_oi_series
    global total_vol_series 
    global maturity_series 
    
    #strike price
    for x in driver.find_elements(By.XPATH, "//tbody/tr/td[9]"): 
        y = float(x.text)/1000
        series1 = pd.Series([y])
        strike_series = pd.concat([strike_series, series1 ])   
    #put prior oi
    for x in driver.find_elements(By.XPATH, "//tbody/tr/td[16]"): 
        y = float(x.text.replace(",",""))
        series1 = pd.Series([y])
        put_prior_oi_series = pd.concat([put_prior_oi_series, series1 ])
    #put est vol
    for x in driver.find_elements(By.XPATH, "//tbody/tr/td[17]"): 
        y = float(x.text.replace(",",""))
        series1 = pd.Series([y])
        put_est_vol_series = pd.concat([put_est_vol_series, series1 ])
    #put oi
    put_oi_series =  put_prior_oi_series + put_est_vol_series 
    #call prior oi
    for x in driver.find_elements(By.XPATH, "//tbody/tr/td[2]"): 
        y = float(x.text.replace(",",""))
        series1 = pd.Series([y])
        call_prior_oi_series = pd.concat([call_prior_oi_series, series1 ])    
    #call est vol
    for x in driver.find_elements(By.XPATH, "//tbody/tr/td[1]"): 
        y = float(x.text.replace(",",""))
        series1 = pd.Series([y])
        call_est_vol_series = pd.concat([call_est_vol_series, series1 ])              
    call_oi_series = call_prior_oi_series + call_est_vol_series 
    total_vol_series = call_oi_series + put_oi_series 
    #maturity date
    len_of_option = range(len(driver.find_elements(By.XPATH, "//tbody/tr/td[9]")))
    for x in len_of_option: 
        series1 = pd.Series(exercise_table[no_of_option])
        maturity_series = pd.concat([maturity_series,series1])
```

```python
strike_series = pd.Series(dtype = 'float64')
put_prior_oi_series = pd.Series(dtype = 'float64')
put_est_vol_series = pd.Series(dtype = 'float64')
put_oi_series = pd.Series(dtype = 'float64')
call_prior_oi_series = pd.Series(dtype = 'float64')
call_est_vol_series = pd.Series(dtype = 'float64')
call_oi_series = pd.Series(dtype = 'float64')
total_vol_series = pd.Series(dtype = 'float64')
maturity_series = pd.Series(dtype = 'float64')


for no_of_option in range(to_execute_on_how_many_options): 
    adam_scrapping(no_of_option)
```

```python
final_dataframe = pd.concat([strike_series,put_oi_series,call_oi_series,total_vol_series,maturity_series],axis=1)
```

```python
filepath = Path(to_store_csv_at) 
filepath.parent.mkdir(parents=True, exist_ok=True)  
final_dataframe.columns=['strike', 'put oi', 'call oi', 'total vol', 'maturity date'] 
final_dataframe.to_csv(filepath,index=False)
```
