# Symmetric Matrix

For PCA, we assume covariance matrix is symmetric or Hermitian because
- covariance matrix is symmetric by definition. covariance between Y and X is the same as the covariance between X and Y

# PCA

```python
corr_matrix = np.corrcoef(df, rowvar=False) 
# False treat column as variable and row as observation (instead of the default: row as variable and col as observation)

eVal,eVec=np.linalg.eigh(corr_matrix) 
#eigen values : eigen value of each "principal compoenent of correlation matrix" is a magnitude that tells us how much variance in entire stock data can be explained by this compoenent, larger eigenvalue means more explantory power
#eigen vectors : eigenvector provide the weights of stocks for each "principal compoenent of correlation matrix"

indices = eVal.argsort() # indices that would sort eVal in ascending order
eVal0, eVec0 = eVal[indices], eVec[:,indices] # reordered according to 'indices', ensuring ascending order of eigenvalues
eVal0 = np.diagflat(eVal0) # convert eVal into a diagonal matrix with eigenvalues on diagonal and zeros elsewhere
```

# PDF of Random Matrix Eigenvalues
we can generate PDF based on Marcenko-Pastur (MP) Law, which is often used in random matrix theory, to analyze eigenvalye distribution of covariance matrices



```python
var = 1. # set variance of matrix element to 1
q = df.shape[0] / float(df.shape[1]) # ratio of rows (observation) to col (variables), i.e. how many observations per stock
pts = 200 # no of points at which PDF will be evaluated 
 
eMin = var*(1-(1./q)**.5)**2 # min eigenvalues
eMax = var*(1+(1./q)**.5)**2 # max eigenvalues 
eVal = np.linspace(eMin, eMax, pts)
pdf0 = q/(2*np.pi*var*eVal)*((eMax-eVal)*(eVal-eMin))**.5 # calc how likely to see an eigenvalue in a perfectly random dataset 
pdf0 = pd.Series(pdf0, index=eVal) # pdf0 is a series where index is eVal, and col is pdf of each eVal
```

# Kernel Density Estimation (KDE)
To estimate PDF of random variable in a non-parametric way



```python
eigenvalues.shape
```

```python
from sklearn.neighbors import KernelDensity
print(eigenvalues.shape)
obs = eigenvalues # observations
bWidth = 0.01 # stepwise for pdf calcualtion
kernel='gaussian'
x = np.unique(obs).reshape(-1,1) # x are the specific points to calc pdf

if len(obs.shape)==1: 
    obs=obs.reshape(-1,1)

if len(obs.shape)==1:
    obs=obs.reshape(-1,1)
    
kde = KernelDensity(kernel=kernel, bandwidth = bWidth).fit(obs)

if len(x.shape) == 1:
    x=x.reshape(-1,1)
    
logProb = kde.score_samples(x)
pdf1 = pd.Series(np.exp(logProb),index=x.flatten())
print(pdf1.shape)
```
