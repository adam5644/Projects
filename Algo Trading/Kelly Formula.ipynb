# Kelly Formula
- a math formula to determine the optimal size of a series of bets
- balance trade-off between return and risk
    - return: maximize compounded growth rate of capital
    - risk: risk of bankcruptcy
- f* is the proportion of your resourecs to bet to maximize your wealth in the long term. If f* is negative, don't bet.
- limitation:
    - wrong estimation of p
        - black swan events?
        - non-gaussian distribution of returns
        - stop loss?



```python
def kelly(p, q, b):
    return (b*p - q)/ b
    
p = 0.6
q = 1-p
b = p/q
fraction_to_bet = kelly(p,q,b)
fraction_to_bet
```
