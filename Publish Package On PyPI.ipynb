# Create __init__.py
- __init__.py file in a python package
    - (Purpose 1) it tells python that the whole directory (Test1) is a package
    - (Purpose 2) it controls what's available when the package is imported

```python
from .main import hello
```

# setup.py
- how to configure and bundle the package

```python
from setuptools import setup, find_packages

setup(
    name='test1', # this is the package folder that contain the "main.py"
    version='0.1', # to increment this version number for each new version of this package
    packages=find_packages(), 
    install_requires=[ # dependencies for this package
    # e.g. 'numpy>=1.11.1'
    ],
)
```

# Command Line Prompt
- Generate source distribution, which is the python script
- Generate wheel distribution, which are platform-specific binaries
- For all the necessary files for installing and running the package


```python
python setup.py sdist bdist_wheel
```
