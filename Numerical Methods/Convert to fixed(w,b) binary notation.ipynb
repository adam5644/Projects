```python
def toFixedPoint(num, w, b):
    res = [0] * w
    if num < 0:
        num += 2**w  # Add 2 to the power of w to handle two's complement
        for i in range(w):
            power = 2**(w-1-i)
            if num >= power:
                num -= power
                res[i] = 1
    else:
        # Handle the case for positive numbers
        for i in range(w):
            power = 2**(w-1-i)
            if num >= power:
                num -= power
                res[i] = 1
    return res

print(toFixedPoint(-10, 8, 1))
```
