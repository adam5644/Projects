# (a)


# European Call Option Pricing using Two-Step Binomial Model

Given:
- Initial stock price $S = $100$
- Strike price $K = $105$
- Risk-free interest rate $r = 3\%$
- Time to maturity $T = 1$ year
- Volatility $\sigma = 20\%$
- Number of steps $n = 2$ for the two-step binomial model

## Step 1: Calculate the time step $(\Delta t)$
$\Delta t = \frac{T}{n} = \frac{1}{2}$ year

## Step 2: Calculate the up and down factors $(u)$ and $(d)$

$u = b + \sqrt{b^2 - 4}$ where $b = e^{rt + \sigma^2 t} + e^{-rt}$

up factor

$d \approx \frac{1}{u}$

down factor

- $u = e^{\sigma \sqrt{\Delta t}} = 1.1547233$
- $d = e^{-\sigma \sqrt{\Delta t}} = 0.86600829$

## Step 3: Calculate the risk-neutral probability $(p)$
$p = \frac{e^{r \Delta t} - d}{u - d} = 0.51644265$

## Step 4: Calculate the option payoff at the end of the tree

- If the stock price goes up twice $(uu)$: 
  $$S_{uu} = S \cdot u^2 = 100 \cdot u^2 = 133.33859$$

- If the stock price goes up then down or vice versa $(ud)$ or $(du)$: 
  $$S_{ud} = S \cdot u \cdot d = S \cdot d \cdot u = 100 \quad (\text{since } ud = 1)$$

- If the stock price goes down twice $(dd)$: 
  $$S_{dd} = S \cdot d^2 = 100 \cdot d^2 = 74.997036$$

## Step 5: Calculate the option value at the initial node

- If the stock price goes up twice $(uu)$:
  $$cuu = \max(S_{uu} - K, 0) = 28.33859$$

- If the stock price goes up then down or vice versa $(ud)$ or $(du)$:
  $$cud = \max(S_{ud} - K, 0) = 0$$

- If the stock price goes down twice $(dd)$:
  $$cdd = \max(S_{dd} - K, 0) = 0$$

- The value of the call option after the first up move $(u)$, discounted back one step:
  $$cu = e^{-r \Delta t} \left( p \cdot cuu + (1 - p) \cdot cud \right)$$
  $$cu = e^{-0.03 \times 0.5} \left( p \cdot 28.33859 \right) = 14.4173659$$

- The current value of the call option, discounted back to today from the two possible outcomes $(u)$ and $(d)$:
  $$c = e^{-0.03 \times 0.5} \left( p \cdot 14.4173659 \right) = 7.33489$$

### Hence,  (a) ans = 7.33489
