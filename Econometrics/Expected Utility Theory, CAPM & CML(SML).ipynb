# Expected utility theory
- to model consumer behavior
**lagrangian objective function**
- a function used in optimization problem to maximize/minimize a function subject to constraints
    - maximize utility function U(X), where x is a vector of goods and U(X) is satisfaction derived from consuming these goods
** lagrange multiplier (λ)**
- a technique to incorporate constraints into the optimization problem
    - λ associated with the budget constraint, λ = marginal utility of income 
    - wealth (Y) = sum(price of x * goods x)
- solution of this optimization problem is 'a set of optimal demands of each goods' which maximize consumer's utility given their budget constraint
    - This is found by setting the partial derivatives of the Lagrangian with respect to each good x and λ to zero and solving the resulting system of equations.
