# Portfolio Immunization
- purpose: ensure that the portfolio's value is not affected by changes in interest rates

- process of obtaining a portfolio of zero dollar duration and dollar convexity
    - Duration measures the sensitivity of a bond's price to changes in interest rates, while convexity measures the rate of change of duration as interest rates change
- Asset-Liability Management
    - Immunization is a key technique in asset-liability management, where it's used to manage the risks that arise from mismatches between assets and liabilities, ensuring that the assets will cover the liabilities at any point in time
