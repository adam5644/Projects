# Interest rate risk in the banking book (IRRBB)
- current or propective risk to bank's capital and earnings arising from adverse movement in interest rates that affect bank's **banking book positions**
- **repricing risk**
    - risk arising from mismatching of reset timing (**repricing date**) of interest rates on asset and liabilities and their exposure amounts (**repricing gap**)
    - **(example)** rates increase and bank's deposit reprice sooner than loans, hence bank pay more interest on deposits than the interest receives from loans. Affecting bank's **net interest income (NII)** and **economic value of equity (EVE)** which is derived by discounting future cash inflow and outflows
    

# determine repricing date and amount (repricing amt +ve if bank receive money)
- fixed rate deposit (loan)
    - 1 million fixed rate deposit (loan) expiring in 2y
    - repricing date = 2y
    - repricing amt = +1mil (-1mil)
- floating rate deposit and loan
    - 1mil deposit expiring in 2y with deposit rate being 3m libor rate, interest paid at end of 3m and 3m libor rate is at start of 3m.
    - repricing date = 3m
    - repricing amt = -1mil
- fixed rate mortgage
    - each loan payment date is repricing date (as bank need to reinvest the principal repayment amt), and repricing amount is corresponding principal repayment amt 
- single-currency fixed-to-floating interest rate swap (irs)
    - irs = 
        - fixed leg as fixed rate deposit (if receive fixed) or loan (if pay fixed)
        - floating as floating rate deposit or loan
- forward starting fixed rate loan
    - Bank has funding need in 3m to raise 50mil loan principal. 
    - **forward starting fixed rate loan = ** 3y fixed rate, 5%, loan and expects to sell 50mil with all draw-down (i.e. under utilization of funds under a loan facility) occuring in 3m. the funding rate for 50mil is not known until 3m later, hence **repricing risk** 
    - repricing risk = -50mil reprciced in 3m (as bank to borrow at rate 3m later)
    - repricing risk = +50 mil reprciing in 3y3m (as bank to return at rate 3y3m later)
