# Pareto Distribution
- has heavy tail to model outliers (e.g. market crash)
- gaussian dist has finite variance
- pareto dist
    - finite variance if alpha >= 2
    - infinite variance if alpha < 2, which means the model has non-zero probability for extremely-large outlier that is not bounded
    

- higher alpha --> less outlier



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto

xm = 1 # scale
alpha_values = [2.05, 3,4] # shape paremeter

x = np.linspace(1,5,1000)
plt.figure(figsize=(12,6))

for alpha in alpha_values:
    pdf = pareto.pdf(x, alpha, scale=xm)
    plt.plot(x, pdf, label=f'a = {alpha}')
    
plt.legend()
```
